## 2. Mulai Training YOLOv8n
Kita menggunakan varian `yolov8n.pt` (Nano) karena ini yang paling optimal untuk dijalankan di NVIDIA Jetson Orin Nano.

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt') # Load pre-trained model

# Sesuaikan path data.yaml dengan folder hasil download Roboflow di atas
# Contoh: data='/content/Baterai-Dataset-1/data.yaml'
results = model.train(
    data='/content/YOUR_DATASET_FOLDER/data.yaml', 
    epochs=100, 
    imgsz=640, 
    batch=16,
    project='recell_vision',
    name='train_1'
)
```
